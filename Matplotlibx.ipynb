# plotting multiple plots


```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green') #coloring the plots 
plt.plot(batsman['index'],batsman['RG Sharma'],color='black') #for coloring i can search on google hex code and select color of my coice
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season') #naming the x axis
plt.ylabel('runs scored') #naming y axis
```

# line styling

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',markersize=10)  
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='o')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season') 
plt.ylabel('runs scored') # in this i just change style of line using linestyle
```

# legend() is used to naming the plot wehther this line belongs to this person or anything using label

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',markersize=10,label= 'V kohli')  
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='o',label= 'R Sharma')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season') 
plt.ylabel('runs scored') 
plt.legend(loc='best') #we can use here upper ,right,left etc to move  kohli & R sharma which are inside the graph
plt.grid() #add grids to your graph
```

```python
#plt.bar(df['batsman'],df['2015'],width=0.5)
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=-0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='blue')
plt.xticks(np.arange(df.shape[0]),df['batsman']) #for writing name again instead of 0,1,2,3,4 
plt.show()
```

## stack bar chart

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016'] + df['2017'],label='2015')
plt.legend()
plt.show()
```
